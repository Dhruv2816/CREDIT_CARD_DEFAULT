# **Default of Credit Card Clients Dataset**

**About Dataset**

Dataset Information

This dataset contains information on default payments, demographic factors, credit data, history of payment, and bill statements of credit card clients in Taiwan from April 2005 to September 2005.

Content
There are 25 variables:

1. ID: ID of each client
2. LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
3. SEX: Gender (1=male, 2=female)
4. EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
5. MARRIAGE: Marital status (1=married, 2=single, 3=others)
6. AGE: Age in years
7. PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, â€¦ 8=payment delay for eight months, 9=payment delay for nine months and above)
8. PAY_2: Repayment status in August, 2005 (scale same as above)
9. PAY_3: Repayment status in July, 2005 (scale same as above)
10. PAY_4: Repayment status in June, 2005 (scale same as above)
11. PAY_5: Repayment status in May, 2005 (scale same as above)
12. PAY_6: Repayment status in April, 2005 (scale same as above)
13. BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)
14. BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
15. BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
16. BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
17. BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
18. BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
19. PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
20. PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
21. PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
22. PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
23. PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
24. PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
25. default.payment.next.month: Default payment (1=yes, 0=no)

# PCA

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
df.rename(columns = {'default.payment.next.month':'Target'}, inplace = True)
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
    return vif_data

X_original = df.drop(columns=['ID', 'Target'])
vif_data = calculate_vif(X_original)
print(vif_data)
```

```python
vif_threshold = 5
features_to_keep = vif_data[vif_data["VIF"] < vif_threshold]["feature"]
X_filtered = X_original[features_to_keep]

print("Features to keep after removing high VIF:")
print(features_to_keep)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filtered)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filtered)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Component')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.axhline(y=0.90, color='r', linestyle='--', label='90% Explained Variance')
plt.axhline(y=0.95, color='g', linestyle='--', label='95% Explained Variance')
plt.legend()
plt.show()

n_components = np.argmax(cumulative_explained_variance >= 0.90) + 1
print(f"Number of components to keep for 90% explained variance: {n_components}")

pca = PCA(n_components=n_components)
X_pca_optimal = pca.fit_transform(X_scaled)

X_pca_df = pd.DataFrame(X_pca_optimal)
X_pca_df.head()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
param_grid_logreg = {
    'C': [0.1, 1, 10],
    'solver': ['liblinear', 'saga']
}

param_grid_rf = {
    'n_estimators': [10, 50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}


```

```python
results = {}
for clf_name, clf in classifiers.items():
    if clf_name != 'Naive Bayes':  # Skip GridSearchCV for Naive Bayes
        param_grid = {}  # Define your parameter grid for GridSearchCV if needed
        grid_search = GridSearchCV(clf, param_grid, cv=5)  # Example: GridSearchCV
        grid_search.fit(X_train, y_train)
        best_clf = grid_search.best_estimator_  # Get the best estimator found by Grid Search
    else:
        best_clf = clf  # Use GaussianNB directly

    best_clf.fit(X_train, y_train)  # Fit the best classifier on the training data

    # Predict using the validation set
    y_val_pred = best_clf.predict(X_val)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    val_precision = precision_score(y_val, y_val_pred, average='weighted')
    val_recall = recall_score(y_val, y_val_pred, average='weighted')
    val_f1 = f1_score(y_val, y_val_pred, average='weighted')
    val_cm = confusion_matrix(y_val, y_val_pred)

    # Predict using the test set
    y_test_pred = best_clf.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred, average='weighted')
    test_recall = recall_score(y_test, y_test_pred, average='weighted')
    test_f1 = f1_score(y_test, y_test_pred, average='weighted')
    test_cm = confusion_matrix(y_test, y_test_pred)

    results[clf_name] = {
        'Validation': {
            'Accuracy': val_accuracy,
            'Precision': val_precision,
            'Recall': val_recall,
            'F1 Score': val_f1,
            'Confusion Matrix': val_cm
        },
        'Test': {
            'Accuracy': test_accuracy,
            'Precision': test_precision,
            'Recall': test_recall,
            'F1 Score': test_f1,
            'Confusion Matrix': test_cm
        },
        'Best Parameters': best_clf.get_params()  # Store best parameters for analysis
    }

# Print or analyze results as needed
for clf_name, result in results.items():
    print(f"Classifier: {clf_name}")
    print("Validation Metrics:")
    print(f"  Accuracy: {result['Validation']['Accuracy']:.4f}")
    print(f"  Precision: {result['Validation']['Precision']:.4f}")
    print(f"  Recall: {result['Validation']['Recall']:.4f}")
    print(f"  F1 Score: {result['Validation']['F1 Score']:.4f}")
    print(f"  Confusion Matrix:\n{result['Validation']['Confusion Matrix']}")
    print("Test Metrics:")
    print(f"  Accuracy: {result['Test']['Accuracy']:.4f}")
    print(f"  Precision: {result['Test']['Precision']:.4f}")
    print(f"  Recall: {result['Test']['Recall']:.4f}")
    print(f"  F1 Score: {result['Test']['F1 Score']:.4f}")
    print(f"  Confusion Matrix:\n{result['Test']['Confusion Matrix']}")
    print(f"Best Parameters: {result['Best Parameters']}")
```

### Neural Networks

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
# Convert the DataFrame to NumPy arrays
X_train_array = X_train.to_numpy()
y_train_array = y_train.to_numpy()
X_test_array = X_test.to_numpy()
y_test_array = y_test.to_numpy()

# Convert the data to PyTorch tensors
X_train_tensor = torch.tensor(X_train_array, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_array, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_array, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_array, dtype=torch.float32)

# Create DataLoader for training and testing sets
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
import torch.nn as nn
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_sizes, dropout_rate):
        super(NeuralNetwork, self).__init__()
        self.layers = nn.ModuleList()
        prev_size = input_size
        for hidden_size in hidden_sizes:
            self.layers.append(nn.Linear(prev_size, hidden_size))
            self.layers.append(nn.ReLU())
            self.layers.append(nn.Dropout(dropout_rate))
            prev_size = hidden_size
        self.output = nn.Linear(prev_size, 1)

    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        x = torch.sigmoid(self.output(x))
        return x
```

```python
param_grid = {
    'hidden_sizes': [[64, 32, 16], [128, 64, 32], [256, 128, 64]],
    'dropout_rate': [0.3, 0.4, 0.5],
    'learning_rate': [0.001, 0.0001],
    'batch_size': [32, 64]
}
```

```python
import torch
import torch.optim as optim
from sklearn.metrics import accuracy_score, precision_score, recall_score

def train_and_evaluate(train_loader, test_loader, input_size, hidden_sizes, dropout_rate, learning_rate, num_epochs=10):
    model = NeuralNetwork(input_size, hidden_sizes, dropout_rate)
    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    train_accuracy, train_precision, train_recall = [], [], []
    val_accuracy, val_precision, val_recall = [], [], []

    for epoch in range(num_epochs):
        model.train()
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs.float()).squeeze()
            loss = criterion(outputs, targets.float())
            loss.backward()
            optimizer.step()

        # Calculate metrics on training set
        model.eval()
        all_preds, all_targets = [], []
        with torch.no_grad():
            for inputs, targets in train_loader:
                outputs = model(inputs.float()).squeeze()
                preds = (torch.sigmoid(outputs) > 0.5).int()
                all_preds.extend(preds.cpu().numpy())
                all_targets.extend(targets.cpu().numpy())
        train_accuracy.append(accuracy_score(all_targets, all_preds))
        train_precision.append(precision_score(all_targets, all_preds, average='macro'))
        train_recall.append(recall_score(all_targets, all_preds, average='macro'))

        # Calculate metrics on validation set
        all_preds, all_targets = [], []
        with torch.no_grad():
            for inputs, targets in test_loader:
                outputs = model(inputs.float()).squeeze()
                preds = (torch.sigmoid(outputs) > 0.5).int()
                all_preds.extend(preds.cpu().numpy())
                all_targets.extend(targets.cpu().numpy())
        val_accuracy.append(accuracy_score(all_targets, all_preds))
        val_precision.append(precision_score(all_targets, all_preds, average='macro'))
        val_recall.append(recall_score(all_targets, all_preds, average='macro'))

    return train_accuracy, train_precision, train_recall, val_accuracy, val_precision, val_recall
```

```python
import torch.optim as optim
import itertools
```

```python
import numpy as np
best_params = None
best_score = 0

best_train_acc, best_train_prec, best_train_rec = None, None, None
best_val_acc, best_val_prec, best_val_rec = None, None, None

for hidden_sizes, dropout_rate, learning_rate, batch_size in itertools.product(
    param_grid['hidden_sizes'], param_grid['dropout_rate'], param_grid['learning_rate'], param_grid['batch_size']):

    # Update the DataLoader with new batch_size
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    train_acc, train_prec, train_rec, val_acc, val_prec, val_rec = train_and_evaluate(
        train_loader, test_loader, X_train.shape[1], hidden_sizes, dropout_rate, learning_rate)

    score = (val_acc[-1] + val_prec[-1] + val_rec[-1]) / 3  # Average score for the last epoch

    if score > best_score:
        best_score = score
        best_params = {
            'hidden_sizes': hidden_sizes,
            'dropout_rate': dropout_rate,
            'learning_rate': learning_rate,
            'batch_size': batch_size
        }
        best_train_acc, best_train_prec, best_train_rec = train_acc, train_prec, train_rec
        best_val_acc, best_val_prec, best_val_rec = val_acc, val_prec, val_rec

# Plot the best learning curves
epochs = range(1, len(best_train_acc) + 1)
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(epochs, best_train_acc, label='Train Accuracy')
plt.plot(epochs, best_val_acc, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Best Learning Curve (Accuracy)')

plt.subplot(1, 3, 2)
plt.plot(epochs, best_train_prec, label='Train Precision')
plt.plot(epochs, best_val_prec, label='Validation Precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()
plt.title('Best Learning Curve (Precision)')

plt.subplot(1, 3, 3)
plt.plot(epochs, best_train_rec, label='Train Recall')
plt.plot(epochs, best_val_rec, label='Validation Recall')
plt.xlabel('Epochs')
plt.ylabel('Recall')
plt.legend()
plt.title('Best Learning Curve (Recall)')

plt.tight_layout()
plt.show()

print(f'Best Parameters: {best_params}')
```
